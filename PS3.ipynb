```python
#Queston1:
#step1,2,3
import pandas as pd
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
users.head(10)
```

```python
#step4 Discover what is the mean age per occupation
grp_by=users.groupby('occupation')

mean_occupation=grp_by.age.mean()

mean_occupation
```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
users[users['gender']=='M'].groupby('occupation')['gender'].count().sort_values(ascending=False)
```

```python
#Step 6 For each occupation, calculate the minimum and maximum ages
occu_age_minmax = users.groupby('occupation').agg({'age':['min', 'max']})
print(occu_age_minmax)
```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age 
print("Mean age by occupation and Gender")
users.groupby(['occupation','gender'])['age'].mean()
```

```python
#Step 8. For each occupation present the percentage of women and men
gen_occu=users.groupby(['occupation','gender']).agg({'gender':'count'})
occup_count = users.groupby(['occupation']).count()
occup_gender = gen_occu.div(occup_count, level = "occupation")*100
occup_gender.loc[:, 'gender']
```

```python
#Step 3
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
#re-arrange cols
data = data[['Date','RPT','VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']]
data.head()
```

```python
#Step 4
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,
data.Date,data.Date - pd.offsets.DateOffset(years=100))
data.head()
```

```python
#Step5
newData = data.set_index("Date")
newData.index.astype("datetime64[ns]")
```

```python
#Step8
x= newData.mean()
x
```

```python
#Step9
def stats(x):
    x = panda.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = panda.Series(res,index=indx)
    return res
loc_stats = newData.apply(stats)
loc_stats
```

```python
#Step10
day_stats = newData.apply(stats,axis=1)
day_stats.head()
```

```python
#Question6
import pandas as panda
import matplotlib.pyplot as pplt
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data.plot(x = 'Year', y = ['Marriages_per_1000','Divorces_per_1000'],
label = ['Marraiges' , 'Divorces'])
pplt.title('Marraiges & Divorces Per Capita', color = 'Yellow', fontsize = 20)
pplt.xlabel('Year', color = 'Blue', fontsize = 20)
pplt.ylabel('Per Capita', color = 'Red', fontsize = 20)
pplt.show()
```

```python
#Question7
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
x=pd.DataFrame(data.loc[(data['Year'] == 1900)])
y=pd.DataFrame(data.loc[(data['Year'] == 1950)])
z=pd.DataFrame(data.loc[(data['Year'] == 2000)])
final = pd.concat([x, y, z])
plt.rcParams["figure.figsize"] = (9, 9)
final.plot(x="Year", y=["Marriages_per_1000", "Divorces_per_1000"], kind="bar")
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000')
plt.show()
```
